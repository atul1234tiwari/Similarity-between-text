```python
import pandas as pd
import spacy
```

```python
data=pd.read_csv("Text_Similarity_Dataset.csv")
data.head()
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
li=[]
for i in range(data.Unique_ID.max()+1):
    t1=nlp(data.text1.iloc[i]) 
    t2=nlp(data.text2.iloc[i])
    li.append(t1.similarity(t2))
```

```python
result=pd.DataFrame({'Unique_ID':data.Unique_ID,'similarity_score':li})
result.to_csv('similarity_score.csv',index=False)
```
